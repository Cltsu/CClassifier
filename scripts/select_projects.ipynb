```python
# 抽取出metadata
import extract_metadata
import os
import pandas as pd
import config

project_path = config.project_path
data_path = config.data_path

result_path = project_path + 'results\\'

repos = []
with open(os.path.join(project_path, 'results\\projects.csv'), 'r', encoding='utf-8') as f:
    repos = f.read().splitlines()

conflict_path = data_path + 'conflicts_files\\'

label_dis = {}

for repo in repos:
    if repo in label_dis.keys():
        continue
    repo_path = conflict_path + repo + '\\'
    # print(repo_path)
    label_dis[repo] = extract_metadata.extract(repo_path)

label_distribution = pd.DataFrame.from_dict(label_dis).transpose()
label_distribution.to_csv(os.path.join(result_path, 'label_distribution_tmp.csv'), index=True)
label_distribution
```

```python
# 去除部分A、B所占比重过大的repo
ld = label_distribution.copy()
ld['CC'] = ld['CC12'] + ld['CC21']
ld = ld.drop(['CC12', 'CC21'], axis=1)
ld[['A','B','CC','CB','NC']] = ld[['A','B','CC','CB','NC']].div(ld['found'], axis=0)


ld = ld.drop(ld[(ld['A'] > 0.6) | (ld['B'] > 0.6) | (ld['A'] + ld['B'] > 0.8) | (ld['found'] < 600)].index )
ld
```

```python
# 标记index并移动
import shutil
import os
import config
# project_path = config.project_path
# data_path = config.data_path

project_path = config.project_path
data_path = config.data_path

repos = []
with open(project_path + 'results\\projects.csv', 'r', encoding='utf-8') as f:
    repos = f.read().splitlines()

conflict_path = data_path + 'conflicts_files\\'
index_path = data_path + 'conflicts_files_index\\'

chunk_index = 0
for repo in repos:
    save_path = index_path + repo + '\\'
    cur_path = conflict_path + repo + '\\'
    os.makedirs(save_path)
    for home, dirs, files in os.walk(cur_path):
        for filename in files:
            if filename.endswith('metadata.json') and os.path.isfile(os.path.join(home, 'metadata.json')) and os.path.isfile(os.path.join(home, 'ours.java')) and os.path.isfile(os.path.join(home, 'theirs.java')) and os.path.isfile(os.path.join(home, 'conflict.java')) and os.path.isfile(os.path.join(home, 'resolve.java')):
                shutil.copy(os.path.join(home, 'metadata.json'), os.path.join(save_path, str(chunk_index) + '_' + 'metadata.json'))
                shutil.copy(os.path.join(home, 'ours.java'), os.path.join(save_path, str(chunk_index) + '_' + 'a.java'))
                shutil.copy(os.path.join(home, 'theirs.java'), os.path.join(save_path, str(chunk_index) + '_' + 'b.java'))
                shutil.copy(os.path.join(home, 'conflict.java'), os.path.join(save_path, str(chunk_index) + '_' + 'merged.java'))
                shutil.copy(os.path.join(home, 'resolve.java'), os.path.join(save_path, str(chunk_index) + '_' + 'resolved.java'))
                if os.path.isfile(os.path.join(home, 'base.java')):
                    shutil.copy(os.path.join(home, 'base.java'), os.path.join(save_path, str(chunk_index) + '_' + 'base.java'))
                chunk_index += 1
```
